```python
import pandas as pd
import numpy as np

# Viz
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import cm
%matplotlib inline
%config Inlinebackend.figure_format = 'retina'
sns.set_context("poster")
sns.set(rc={"figure.figsize": (15.,8.)})
sns.set_style("whitegrid")

import plotly.express as px
import plotly.io as pio
```

```python
TP_DATA = pd.read_csv('../DATA/turbine_Performance_prediction_dataframe.csv')
TP_DATA.drop(columns=['WindSpeed (m/s)', 'Wind Direction (°)', 'Unnamed: 0'], inplace=True)
TP_DATA['Time'] = pd.to_datetime(TP_DATA['Time'])
TP_DATA
```

```python
# Convert the 'Time' column to a datetime type
TP_DATA['Time'] = pd.to_datetime(TP_DATA['Time'])

# Create the line plot
fig = px.line(TP_DATA, x='Time', y='Total Active Power (kWh)')

# Show the plot
fig.show()
```

```python
# Export the plot as a URL
url = fig.to_html(full_html=False)
print(url)
```
